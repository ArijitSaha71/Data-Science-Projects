```python
import numpy as np
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.manifold import MDS
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
X_unsup = np.load("/content/drive/MyDrive/data/p2_unsupervised/X.npy")
X_unsup.shape
```

```python
X_unsup_log = np.log2(X_unsup+1)
```

```python
np.amax(X_unsup_log[:,0])
```

```python
X_pca = PCA(n_components=1600, svd_solver='full') # Initialize with n_components parameter to only find the top eigenvectors
Z_pca = X_pca.fit_transform(X_unsup_log)
```

```python
np.where(np.cumsum(X_pca.explained_variance_ratio_) >=.85)[0][0]
```

```python
np.where(np.cumsum(X_pca.explained_variance_ratio_) >=.40)[0][0]
```

```python
Z_85 = Z_pca[:,0:1252]
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.scatter(Z_pca[:,0],Z_pca[:,1])
plt.title("Scatter plot of 1st and 2nd PCs",size=18)
plt.xlabel("PC 1",size=14)
plt.ylabel("PC 2",size=14)
#plt.axis("equal")
plt.show()
```

```python
X_tsne = TSNE(n_components=2,verbose=1,perplexity=40,n_jobs=-1)
Z_tsne = X_tsne.fit_transform(Z_85)
plt.figure(figsize=(6, 4), dpi=100)
plt.scatter(Z_tsne[:,0],Z_tsne[:,1])
plt.title("TSNE, perplexity 40",size=18)
#plt.axis("equal")
plt.show()
```

```python
perplexities = [15, 20, 30, 40, 50, 60]
learning_rates = [150, 200, 250, 300]
for perplexity in perplexities:
  for learning_rate in learning_rates:
    tsne_i = TSNE(n_components=2,perplexity=perplexity, learning_rate=learning_rate,n_jobs=-1)
    tsne_out = tsne_i.fit_transform(Z_85)
    plt.figure(figsize=(4, 3), dpi=100)
    plt.scatter(tsne_out[:,0],tsne_out[:,1])
    plt.title(f"TSNE, perplexity={perplexity}, L_rate={learning_rate}",size=18)
    plt.show()
```

```python
plt.figure(figsize=(6, 4), dpi=100)
X_mds = MDS(n_components=2,verbose=1,eps=1e-5,n_init=4)
Z_mds = X_mds.fit(X_unsup_log)
plt.scatter(Z_mds.embedding_[:,0],Z_mds.embedding_[:,1])
plt.title("MDS Plot",size=18)
plt.show()
```

```python
all_kmeans = [KMeans(n_clusters=i+6,n_init=10) for i in range(10)]
# i-th kmeans fits i+2 clusters
for i in range(10):
    all_kmeans[i].fit(Z_85)

inertias = [all_kmeans[i].inertia_ for i in range(10)]
plt.figure(figsize=(6, 4), dpi=100)
plt.plot(np.arange(6,16),inertias)
plt.title("KMeans Sum of Squares Criterion",size=18)
plt.xlabel("# Clusters",size=14)
plt.ylabel("Within-Cluster Sum of Squares",size=14)
plt.show()
```

```python
kmeans_14 = all_kmeans[8]
kmeans_pca_14 = kmeans_14.predict(Z_85)
plt.figure(figsize=(6, 4), dpi=100)
plt.scatter(Z_85[:,0],Z_85[:,1],c=kmeans_pca_14)
plt.title("KMeans Clustering K=14, PCA Plot",size=18)
plt.xlabel("PC 1",size=14)
plt.ylabel("PC 2",size=14)
plt.show()
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.scatter(Z_tsne[:,0],Z_tsne[:,1],c=kmeans_pca_14)
plt.title("KMeans Clustering K=14, TSNE Plot",size=18)
plt.xlabel("PC 1",size=14)
plt.ylabel("PC 2",size=14)
plt.show()
```

```python
kmeans_13 = all_kmeans[7]
kmeans_pca_13 = kmeans_13.predict(Z_85)
plt.figure(figsize=(6, 4), dpi=100)
plt.scatter(Z_85[:,0],Z_85[:,1],c=kmeans_pca_13)
plt.title("KMeans Clustering K=13, PCA Plot",size=18)
plt.xlabel("PC 1",size=14)
plt.ylabel("PC 2",size=14)
plt.show()
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.scatter(Z_tsne[:,0],Z_tsne[:,1],c=kmeans_pca_13)
plt.title("KMeans Clustering K=13, TSNE Plot",size=18)
plt.xlabel("PC 1",size=14)
plt.ylabel("PC 2",size=14)
plt.show()
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.scatter(Z_mds.embedding_[:,0],Z_mds.embedding_[:,1],c=kmeans_pca_13)
plt.title("KMeans Clustering K=13, MDS Plot",size=18)
plt.xlabel("Dim 1",size=14)
plt.ylabel("Dim 2",size=14)
plt.show()
```

```python
from scipy.cluster.hierarchy import ward, cophenet
from scipy.spatial.distance import pdist, squareform
```

```python
Z = pdist(Z_85)
X = ward(Z)
C = cophenet(X)
```

```python
np.corrcoef(Z, C)
```

As Cophenetic Correlation Coefficient is 0.724, the clustering is quite good.
CCC takes values in (-1,1), higher the value better the clustering.

```python
from sklearn.metrics import davies_bouldin_score
# Due to the way it is defined, as a function of the ratio of the within cluster scatter, 
#to the between cluster separation, a lower value will mean that the clustering is better.
```

```python
davies_score_kmeans = []
for i in range(10):
    davies_score_kmeans.append(davies_bouldin_score(Z_85, all_kmeans[i].predict(Z_85)))
```

```python
plt.figure(figsize=(6, 4), dpi=100)
plt.plot(np.arange(6,16),davies_score_kmeans)
plt.title("KMearns Clust. Davies-Bouldin Score",size=18)
plt.xlabel("# Clusters",size=14)
plt.ylabel("Davies-Bouldin Score",size=14)
plt.show()
```

```python
davies_score_gaussmix = []
for i in range(10):
    gmm = GaussianMixture(n_components=i+6, covariance_type='full', tol=0.001, reg_covar=1e-06, n_init=10)
    davies_score_gaussmix.append(davies_bouldin_score(Z_85, gmm.fit_predict(Z_85)))

plt.figure(figsize=(6, 4), dpi=100)
plt.plot(np.arange(6,16),davies_score_gaussmix)
plt.title("Gaussian Mixture Model Davies-Bouldin Score",size=18)
plt.xlabel("# Clusters",size=14)
plt.ylabel("Davies-Bouldin Score",size=14)
plt.show()    
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
```

```python
np.random.seed(317)
perm = np.random.permutation(X_unsup_log.shape[0])
n_train = int(4/5*X_unsup_log.shape[0])
print(n_train)
X_train = X_unsup_log[perm[:n_train]]
y_train = GausMix_out_13[perm[:n_train]]
X_test = X_unsup_log[perm[n_train:]]
y_test = GausMix_out_13[perm[n_train:]]
```

```python
log_reg_suboptimal = LogisticRegression(solver="saga",C=1,penalty="elasticnet",max_iter=100,multi_class="multinomial",l1_ratio=0.7,n_jobs=-1).fit(X_train,y_train)
```

```python
log_reg_suboptimal.score(X_train,y_train)
```

```python
log_reg_suboptimal.score(X_test,y_test)
```

```python
log_reg_suboptimal.coef_.shape
```

```python
# sum of absolute valuse of coeffs over classes
log_reg_coef_sum_over_classes = np.sum(np.abs(log_reg_suboptimal.coef_), axis = 0)
```
